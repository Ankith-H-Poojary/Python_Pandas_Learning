# 8.How to fill null values with median and mode 

```python
df.fillna(5)
```

```python
 for i in df.columns:
        if df[i].dtypes!="object":
            df[i].fillna(df[i].median(),inplace=True)
            
```

```python
for i in df.columns:
        if df[i].dtypes=="object":
            df[i].fillna(df[i].mode()[0],inplace=True)
            
```

```python
df.isnull().sum()
```

# 9.How to get stastistical summary

```python
summary=df.describe().T
summary["IQR"]=summary["75%"]-summary["25%"]
summary["lw"]=summary["25%"]-1.5*summary["IQR"]
summary["uw"]=summary["75%"]+1.5*summary["IQR"]
summary["skew"]=df.select_dtypes(include="number").skew()
summary["kurtosis"]=df.select_dtypes(include="number").kurt()
summary["variance"]=summary["std"]**2
summary["Range"]=summary["max"]-summary["min"]
summary["cv"]=summary["std"]/summary["mean"]
summary
```

# 11.How to merge the series to data frame based on index


```python
nul=pd.Series(df.select_dtypes(include="number").isnull().sum(),name="null_values")
```

```python
nul
```

```python
pd.merge(summary,nul,left_index=True,right_index=True)
```
